```python
#Problem 2
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
%matplotlib inline
np.random.seed(42)
```

```python
N = 3
mu = 1.0
sigma = 0.2 
sample = norm(mu,sigma).rvs(N)
print(sample)

muGrid = np.linspace(0,2,1000)
L1 = norm.pdf(muGrid,loc=sample[0],scale=sigma)
L2 = norm.pdf(muGrid,loc=sample[1],scale=sigma)
L3 = norm.pdf(muGrid,loc=sample[2],scale=sigma)
L = L1*L2*L3 
fig, ax = plt.subplots(figsize=(7, 5))
plt.plot(muGrid, L1, ls='-', c='green', label=r'$L(x_1)$')
plt.plot(muGrid, L2, ls='-', c='red', label=r'$L(x_2)$')
plt.plot(muGrid, L3, ls='-', c='blue', label=r'$L(x_3)$')
plt.plot(muGrid, L, ls='-', c='black', label=r'$L(\{x\})$')

plt.xlim(0.2, 1.8)
plt.ylim(0, 8.0)
plt.xlabel('$\mu$') #Leave out or adjust if no latex
plt.ylabel(r'$p(x_i|\mu,\sigma)$') #Leave out or adjust if no latex
plt.title('MLE for Gaussian Distribution')
plt.legend()

```

```python
N_new = 10
mu = 1.0
sigma = 0.2
sample_mu = []
while len(sample_mu) < 10:
    sample = norm(mu,sigma).rvs(1)
    if sample >= 1.1:
        sample_mu.append(sample)

muGrid = np.linspace(0,3,1000)

L1 = norm.pdf(muGrid,loc=sample_mu[0],scale=sigma)
L2 = norm.pdf(muGrid,loc=sample_mu[1],scale=sigma)
L3 = norm.pdf(muGrid,loc=sample_mu[2],scale=sigma)
L4 = norm.pdf(muGrid,loc=sample_mu[3],scale=sigma)
L5 = norm.pdf(muGrid,loc=sample_mu[4],scale=sigma)
L6 = norm.pdf(muGrid,loc=sample_mu[5],scale=sigma)
L7 = norm.pdf(muGrid,loc=sample_mu[6],scale=sigma)
L8 = norm.pdf(muGrid,loc=sample_mu[7],scale=sigma)
L9 = norm.pdf(muGrid,loc=sample_mu[8],scale=sigma)
L10 = norm.pdf(muGrid,loc=sample_mu[9],scale=sigma)
L = L1*L2*L3*L4*L5*L6*L7*L8*L9*L10
fig, ax = plt.subplots(figsize=(7, 5))
plt.plot(muGrid, L1, ls='-', c='green', label=r'$L(x_1)$')
plt.plot(muGrid, L2, ls='-', c='red', label=r'$L(x_2)$')
plt.plot(muGrid, L3, ls='-', c='blue', label=r'$L(x_3)$')
plt.plot(muGrid, L4, ls='-', c='green', label=r'$L(x_4)$')
plt.plot(muGrid, L5, ls='-', c='red', label=r'$L(x_5)$')
plt.plot(muGrid, L6, ls='-', c='blue', label=r'$L(x_6)$')
plt.plot(muGrid, L7, ls='-', c='green', label=r'$L(x_7)$')
plt.plot(muGrid, L8, ls='-', c='red', label=r'$L(x_8)$')
plt.plot(muGrid, L9, ls='-', c='blue', label=r'$L(x_9)$')
plt.plot(muGrid, L10, ls='-', c='blue', label=r'$L(x_10)$')
plt.plot(muGrid, L, ls='-', c='black', label=r'$L(\{x\})$')

plt.xlim(0.2, 3)
plt.ylim(0, 8.0)
plt.xlabel('$\mu$') #Leave out or adjust if no latex
plt.ylabel(r'$p(x_i|\mu,\sigma)$') #Leave out or adjust if no latex
plt.title('MLE for Gaussian Distribution')
plt.legend()
```

By forcing the sampled values to be greater than 1.1, the resulting gaussians plotted with said chosen values as their mean will be much more tightly clustered. Therefore, the combination of the gaussians will be substantially taller compared to the previous scenario.

```python
#Problem 3
clear = 1
cloudy = 0

N = 10000
def weather(N, today_weather):
    weather_log = []
    day_total = []
    day_count = 0
    today = today_weather
    for i in range(N):
        prob = np.random.random(1)
        if today == cloudy:
            if prob < 0.5:
                tomorrow = cloudy
            else:
                tomorrow = clear
        if today == clear:
            if prob < 0.1:
                tomorrow = cloudy
            else:
                tomorrow = clear

        today = tomorrow
        weather_log.append(today)
        day_total.append(day_count)
        day_count += 1
    return weather_log, day_total
weather_log, days = weather(N, cloudy)
print weather_log
```

```python
plt.figure(figsize=(7,7))
plt.plot(days, weather_log)
plt.show()
```
