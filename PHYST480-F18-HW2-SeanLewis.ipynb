```python
#Problem 2
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm
%matplotlib inline
np.random.seed(42)
```

```python
N = 3
mu = 1.0
sigma = 0.2 
sample = norm(mu,sigma).rvs(N)
print(sample)

muGrid = np.linspace(0,2,1000)
L1 = norm.pdf(muGrid,loc=sample[0],scale=sigma)
L2 = norm.pdf(muGrid,loc=sample[1],scale=sigma)
L3 = norm.pdf(muGrid,loc=sample[2],scale=sigma)
L = L1*L2*L3 
fig, ax = plt.subplots(figsize=(7, 5))
plt.plot(muGrid, L1, ls='-', c='green', label=r'$L(x_1)$')
plt.plot(muGrid, L2, ls='-', c='red', label=r'$L(x_2)$')
plt.plot(muGrid, L3, ls='-', c='blue', label=r'$L(x_3)$')
plt.plot(muGrid, L, ls='-', c='black', label=r'$L(\{x\})$')

plt.xlim(0.2, 1.8)
plt.ylim(0, 8.0)
plt.xlabel('$\mu$') #Leave out or adjust if no latex
plt.ylabel(r'$p(x_i|\mu,\sigma)$') #Leave out or adjust if no latex
plt.title('MLE for Gaussian Distribution')
plt.legend()

```

By forcing the sampled values to be greater than 1.1, the resulting gaussians plotted with said chosen values as their mean will be much more tightly clustered. Therefore, the combination of the gaussians will be substantially taller compared to the previous scenario.

```python
#Problem 3
clear = 1
cloudy = 0

N = 10000
def weather(N, today_weather):
    weather_log = []
    day_total = []
    day_count = 1
    clear_prob_trace = []
    today = today_weather
    for i in range(N):
        prob = np.random.random(1)
        if today == cloudy:
            if prob < 0.5:
                tomorrow = cloudy
            else:
                tomorrow = clear
        if today == clear:
            if prob < 0.1:
                tomorrow = cloudy
            else:
                tomorrow = clear

        today = tomorrow
        weather_log.append(today)
        day_total.append(day_count)
        total_clear = np.sum(weather_log)
        clear_prob = 1.*total_clear/day_count
        clear_prob_trace.append(clear_prob)
        day_count += 1
    return clear_prob_trace, day_total
clear_prob_trace, days = weather(N, clear)
```

```python
clear_prob_trace, days = weather(N, clear)
clear_prob_trace2, days2 = weather(N, cloudy)
plt.figure(figsize=(7,7))
plt.title('10k Days Starting with Clear and Cloudy')
plt.ylabel('Probability of being Clear')
plt.xlabel('Days')
plt.plot(days, clear_prob_trace, label='Start Clear')
plt.plot(days2, clear_prob_trace2, label='Start Cloudy')
plt.legend()
plt.show()
```

```python
clear_prob_trace3, days = weather(N+1, clear)
clear_prob_trace4, days = weather(N+1, cloudy)
print 'Probability of Clear on day N, N+1 when starting clear: {:.3f}, '.format(clear_prob_trace[-1]) \
    + '{:.3f}'.format(clear_prob_trace3[-1])
print 'Probability of Clear on day N, N+1 when starting cloudy: {:.3f}, '.format(clear_prob_trace2[-1]) \
    + '{:.3f}'.format(clear_prob_trace4[-1])
```
